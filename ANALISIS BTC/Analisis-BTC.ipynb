Analisis del BTC 
desde 2020 hasta 2023: Comprender las tendencia desde hace 2 años atrás. 

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')
```

```python
#Importar los Datos y visualizar la cabezera
datos = pd.read_csv('BTC-USD23012023.csv', encoding = "ISO-8859-1")
datos.head()
```

```python
type(datos)
```

```python
#Graficar los datdos
plt.figure(figsize= (10,5))
plt.plot(datos['Close'], label = 'BITCOIN')
plt.title('Precio del Bitcoin desde 2020 hasta 2023')
plt.xlabel('01 Enero 2020 - 23 de Enero 2023')
plt.ylabel('Precio de Cierre ($)')
plt.legend('BITCOIN')
plt.show()
```

# Creacion de medias moviles

```python
MVS30 = pd.DataFrame()
MVS30['Close'] = datos['Close'].rolling(window = 30).mean()
MVS30.head()
```

```python
MVS100 = pd.DataFrame()
MVS100['Close'] = datos['Close'].rolling(window = 100).mean()
```

```python
plt.figure(figsize= (15,10))
plt.plot(datos['Close'], label = 'BITCOIN')
plt.plot(MVS30['Close'], label = 'Media movil simple de 30')
plt.plot(MVS100['Close'], label = 'Media movil simple de 100')
plt.title('Precio del Bitcoin desde 2020 hasta 2023')
plt.xlabel('01 Enero 2020 - 23 de Enero 2023')
plt.ylabel('Precio de Cierre ($)')
plt.legend(loc = 'upper left')
plt.show()
```
